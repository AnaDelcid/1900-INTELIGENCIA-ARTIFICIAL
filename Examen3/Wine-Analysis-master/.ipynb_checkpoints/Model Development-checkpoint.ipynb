```python
dfq1fsd = df.loc[ : , ["free sulfur dioxide","type"] ].groupby( ["type"] , as_index = False ).quantile(0.25)
dfq1fsd.columns = ["type","free sulfur dioxide_q1"]
```

```python
dfq3fsd =df.loc[ : , ["free sulfur dioxide","type"] ].groupby( ["type"] , as_index = False ).quantile(0.75)
dfq3fsd.columns = ["type","free sulfur dioxide_q3"]
```

```python
dfq2fsd =df.loc[ : , ["free sulfur dioxide","type"] ].groupby( ["type"] , as_index = False ).quantile(0.5)
dfq2fsd.columns = ["type","free sulfur dioxide_q2"]
```

```python
df_iqrfsd = dfq1fsd.merge( dfq3fsd , on=["type"] , how="inner" ).merge( dfq2fsd , on=["type"] , how="inner" )
```

```python
df_iqrfsd["iqr"] = df_iqrfsd.apply( lambda r: r["free sulfur dioxide_q3"] - r["free sulfur dioxide_q1"] , axis = 1)
```

```python
df_iqrfsd["upper"] = df_iqrfsd.apply( lambda r: ( r["free sulfur dioxide_q3"] + 1.5*r["iqr"] ) , axis = 1)
df_iqrfsd["lower"] = df_iqrfsd.apply( lambda r: ( r["free sulfur dioxide_q1"] - 1.5*r["iqr"] ) , axis = 1)
```

```python
df_fsd = df.merge( df_iqrfsd , on=["type"] , how="left" ).copy()
```

```python
def getRealfsd(r):
    if ( r["free sulfur dioxide"] <= r["lower"] ) | ( r["free sulfur dioxide"] >= r["upper"] ):
        return r["free sulfur dioxide_q2"]
    else:
        return r["free sulfur dioxide"]
```

```python
df_fsd.apply( lambda r: getRealfsd(r) , axis = 1 )
```

```python
df_fsd["new_free_sulfur_dioxide"] = df_fsd.apply( lambda r: getRealfsd(r) , axis = 1 )
```

```python
dfq1tsd = df.loc[ : , ["total sulfur dioxide","type"] ].groupby( ["type"] , as_index = False ).quantile(0.25)
dfq1tsd.columns = ["type","total sulfur dioxide_q1"]
dfq3tsd =df.loc[ : , ["total sulfur dioxide","type"] ].groupby( ["type"] , as_index = False ).quantile(0.75)
dfq3tsd.columns = ["type","total sulfur dioxide_q3"]
dfq2tsd =df.loc[ : , ["total sulfur dioxide","type"] ].groupby( ["type"] , as_index = False ).quantile(0.5)
dfq2tsd.columns = ["type","total sulfur dioxide_q2"]

df_iqrtsd = dfq1tsd.merge( dfq3tsd , on=["type"] , how="inner" ).merge( dfq2tsd , on=["type"] , how="inner" )
df_iqrtsd["iqr"] = df_iqrtsd.apply( lambda r: r["total sulfur dioxide_q3"] - r["total sulfur dioxide_q1"] , axis = 1)
df_iqrtsd["upper"] = df_iqrtsd.apply( lambda r: ( r["total sulfur dioxide_q3"] + 1.5*r["iqr"] ) , axis = 1)
df_iqrtsd["lower"] = df_iqrtsd.apply( lambda r: ( r["total sulfur dioxide_q1"] - 1.5*r["iqr"] ) , axis = 1)
df_tsd = df.merge( df_iqrtsd , on=["type"] , how="left" ).copy()
def getRealtsd(r):
    if ( r["total sulfur dioxide"] <= r["lower"] ) | ( r["total sulfur dioxide"] >= r["upper"] ):
        return r["total sulfur dioxide_q2"]
    else:
        return r["total sulfur dioxide"]
df_tsd.apply( lambda r: getRealtsd(r) , axis = 1 )
df_tsd["new_total_sulfur_dioxide"] = df_tsd.apply( lambda r: getRealtsd(r) , axis = 1 )
```

```python
dfq1d = df.loc[ : , ["density","type"] ].groupby( ["type"] , as_index = False ).quantile(0.25)
dfq1d.columns = ["type","density_q1"]
dfq3d =df.loc[ : , ["density","type"] ].groupby( ["type"] , as_index = False ).quantile(0.75)
dfq3d.columns = ["type","density_q3"]
dfq2d =df.loc[ : , ["density","type"] ].groupby( ["type"] , as_index = False ).quantile(0.5)
dfq2d.columns = ["type","density_q2"]

df_iqrd = dfq1d.merge( dfq3d , on=["type"] , how="inner" ).merge( dfq2d , on=["type"] , how="inner" )
df_iqrd["iqr"] = df_iqrd.apply( lambda r: r["density_q3"] - r["density_q1"] , axis = 1)
df_iqrd["upper"] = df_iqrd.apply( lambda r: ( r["density_q3"] + 1.5*r["iqr"] ) , axis = 1)
df_iqrd["lower"] = df_iqrd.apply( lambda r: ( r["density_q1"] - 1.5*r["iqr"] ) , axis = 1)
df_d = df.merge( df_iqrd , on=["type"] , how="left" ).copy()
def getReald(r):
    if ( r["density"] <= r["lower"] ) | ( r["density"] >= r["upper"] ):
        return r["density_q2"]
    else:
        return r["density"]
df_d.apply( lambda r: getReald(r) , axis = 1 )
df_d["new_density"] = df_d.apply( lambda r: getReald(r) , axis = 1 )
```

```python
dfq1a = df.loc[ : , ["alcohol","type"] ].groupby( ["type"] , as_index = False ).quantile(0.25)
dfq1a.columns = ["type","alcohol_q1"]
dfq3a =df.loc[ : , ["alcohol","type"] ].groupby( ["type"] , as_index = False ).quantile(0.75)
dfq3a.columns = ["type","alcohol_q3"]
dfq2a =df.loc[ : , ["alcohol","type"] ].groupby( ["type"] , as_index = False ).quantile(0.5)
dfq2a.columns = ["type","alcohol_q2"]

df_iqra = dfq1a.merge( dfq3a , on=["type"] , how="inner" ).merge( dfq2a , on=["type"] , how="inner" )
df_iqra["iqr"] = df_iqra.apply( lambda r: r["alcohol_q3"] - r["alcohol_q1"] , axis = 1)
df_iqra["upper"] = df_iqra.apply( lambda r: ( r["alcohol_q3"] + 1.5*r["iqr"] ) , axis = 1)
df_iqra["lower"] = df_iqra.apply( lambda r: ( r["alcohol_q1"] - 1.5*r["iqr"] ) , axis = 1)
df_a = df.merge( df_iqra , on=["type"] , how="left" ).copy()
def getReala(r):
    if ( r["alcohol"] <= r["lower"] ) | ( r["alcohol"] >= r["upper"] ):
        return r["alcohol_q2"]
    else:
        return r["alcohol"]
df_a.apply( lambda r: getReala(r) , axis = 1 )
df_a["new_alcohol"] = df_a.apply( lambda r: getReala(r) , axis = 1 )
```

```python
dfq1q = df.loc[ : , ["quality","type"] ].groupby( ["type"] , as_index = False ).quantile(0.25)
dfq1q.columns = ["type","quality_q1"]
dfq3q =df.loc[ : , ["quality","type"] ].groupby( ["type"] , as_index = False ).quantile(0.75)
dfq3q.columns = ["type","quality_q3"]
dfq2q =df.loc[ : , ["quality","type"] ].groupby( ["type"] , as_index = False ).quantile(0.5)
dfq2q.columns = ["type","quality_q2"]

df_iqrq = dfq1q.merge( dfq3q , on=["type"] , how="inner" ).merge( dfq2q , on=["type"] , how="inner" )
df_iqrq["iqr"] = df_iqrq.apply( lambda r: r["quality_q3"] - r["quality_q1"] , axis = 1)
df_iqrq["upper"] = df_iqrq.apply( lambda r: ( r["quality_q3"] + 1.5*r["iqr"] ) , axis = 1)
df_iqrq["lower"] = df_iqrq.apply( lambda r: ( r["quality_q1"] - 1.5*r["iqr"] ) , axis = 1)
df_q = df.merge( df_iqrq , on=["type"] , how="left" ).copy()
def getRealq(r):
    if ( r["quality"] <= r["lower"] ) | ( r["quality"] >= r["upper"] ):
        return r["quality_q2"]
    else:
        return r["quality"]
df_q.apply( lambda r: getRealq(r) , axis = 1 )
df_q["new_quality"] = df_q.apply( lambda r: getRealq(r) , axis = 1 )
```

```python
##Regresion Logistica
```

```python
df_eq['ntype'] = df_eq['type'].apply(lambda x: 0 if x == "white" else 1)
```

```python
df_eq=df_eq.fillna(0)
df_eq
```

```python
features_col = df_eq.columns.values[1:7]
features_col
```

```python
X = df_eq[features_col].values
y = df_eq["respuesta"].values
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,y,test_size = 0.25, random_state=0)
```

```python
log_model = LogisticRegression(solver='lbfgs', max_iter=1000)
log_model.fit(X_train,Y_train)
```

```python
y_pred  = log_model.predict(X_test)
```

```python
log_model.predict_proba(X_test)[:] 
```

```python
cnf_matrix = metrics.confusion_matrix(Y_test, y_pred)
```

```python
import numpy as np
class_names = [0,1]
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)

sns.heatmap(pd.DataFrame(cnf_matrix), annot = True, cmap="Blues_r", fmt="g")
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title("Matriz de confusion", y=1.1)
plt.ylabel("Etiqueta Actual")
plt.xlabel("Etiqueta de prediccion")
```

```python
print(metrics.accuracy_score(Y_test,y_pred))
print(precision_score(Y_test, y_pred))
print(recall_score(Y_test, y_pred))
print(f1_score(Y_test, y_pred))
```

```python
metrics.roc_auc_score(Y_test, y_pred)
```
