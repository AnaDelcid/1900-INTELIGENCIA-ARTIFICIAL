```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("unemployment analysis.csv", delimiter=",")
```

```python
df.dtypes
```

```python
df.describe()
```

```python
dffinal = pd.DataFrame()
for y in df.columns[ 2 :  ]:
    dftemp  = df.loc[ :  , ['Country Name','Country Code', y] ]
    dftemp['Year'] = y
    dftemp.columns = ['Country Name', 'Country Code', 'Value', 'Year']
    dftemp = dftemp.loc[ : , ['Country Name', 'Country Code', 'Year', 'Value'] ]
    
    dffinal = pd.concat([dffinal, dftemp])
```

```python
dffinal
```

```python
dc = pd.read_csv("countries of the world.csv", delimiter=",")
dcw=pd.DataFrame()
dcw['Country Name']= dc.Country
dcw['Population']= dc.Population
dcw
```

```python
dffinal.merge(dcw, on='Country Name', how='left')
```
